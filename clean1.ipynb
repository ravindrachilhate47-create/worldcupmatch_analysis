```python
import pandas as pd
```

```python
df=pd.read_csv("./new_bill.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df["First_Name"]=df["First_Name"].str.capitalize()
df["Last_Name"]=df["Last_Name"].str.capitalize()
```

```python
df["Age"]=pd.to_numeric(df["Age"],errors='coerce')
df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Age"]=df["Age"].astype('int64')
```

```python
df["Age"]
```

```python
df["Email"]=df["Email"].str.lower()
df["Email"]=df["Email"].str.replace(".","")
```

```python
df["Email"]
```

```python
df['Email'] = df['Email'].str.replace('rediffmailcom', 'rediffmail.com', regex=False)
```

```python
df["Email"]=df["Email"].str.replace("invalid_email_no_at","invelide")
df["Email"]=df["Email"].str.replace("NaN","invelide")
```

```python
df["Email"]
```

```python
df["Phone"]=pd.to_numeric(df["Phone"],errors="coerce")
df["Phone"]=df["Phone"].astype('Int64')
```

```python
df["Phone"]
```

```python
df["City"]=df['City'].str.upper()
```

```python
df.info()
```

```python
df["Gender"]=df["Gender"].str.lower()
df["Gender"]=df["Gender"].str.replace('m',"male")
df["Gender"]=df["Gender"].str.replace('f',"female")
```

```python
df["Gender"]
```

```python
df.info()
```

```python
df["Department"]=df["Department"].str.title()
```

```python
df.info()
```

```python
df["Job_Title"]=df["Job_Title"].str.capitalize()
```

```python
df["Payment_Method"]
```

```python
df["Salary"]=pd.to_numeric(df["Salary"],errors="coerce")
df["Salary"].astype("Int64")
```

```python
df["Payment_Method"]=df["Payment_Method"].str.upper()
```

```python
df["Purchase_Date"]=pd.to_datetime(df["Purchase_Date"],format="mixed")
```

```python
df["Rating"]=pd.to_numeric(df["Rating"],errors="coerce")
df["Rating"]=df["Rating"].astype("Int64")
```

```python
df["Status"]=df["Status"].str.upper()
```

```python
df.info()
```

```python
df["Salary"]
```

```python
df.to_csv("new_clean_billbook.csv")
```

```python
df["Payment_Amount"]=df["Payment_Amount"].str.replace("$","")
df["Payment_Amount"]=df["Payment_Amount"].str.replace(".","")
df["Payment_Amount"]=df["Payment_Amount"].str.replace(",","")
```
