```python
#  first is employee column now  this was not null values 
df['employee_id']
```

```python
# this column a all worlds to capitalize now()
df['name']=df['name'].str.capitalize()
```

```python
# so this column a solve the nan values using fillna  
df['email']=df['email'].fillna ("unknown")
```

```python
# Coerced column data type to numeric to resolve formatting errors
# Calculated the mean of the column, ignoring existing NaNs
# Filled all NaN values with the calculated mean 
# and change the data type 
df['age']=pd.to_numeric(df['age'],errors='coerce')
df['age']=df['age'].fillna(df['age'].mean())
df['age'].astype("int64")
```

```python
# this column using a column data type to numeric to resolve formatting errors
# Calculated the mean of the column  and change the data type 
df['salary']=pd.to_numeric(df["salary"],errors='coerce')
df['salary']=df['salary'].fillna(df['salary'].mean())
df['salary']=df['salary'].astype("int64")
```

```python
df['city']=df['city'].str.capitalize()
```

```python
df['city']=df['city'].fillna('unknown ')
```

```python
# fixed the join_date data type using function datetime
df['join_date']=pd.to_datetime(df['join_date'],format='mixed',errors='coerce')
```

```python
# This looks for the exact match of 'Y' or 'N' and replaces it completely
df['active']=df['active'].replace({'Y': 'YES', 'N': 'NO'})
```

```python
df.to_csv("clean_massy.csv", index=False)
```
