```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Ravindra Chilhate\Downloads\employee_data_missing.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Name']=df['Name'].str.strip()
df['Name']=df['Name'].str.upper()
```

```python
df['Department']=df['Department'].fillna("Unknown")                                      
```

```python
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Age']=df['Age'].astype("int64")
```

```python
df['City']=df['City'].str.capitalize()
```

```python
df['Salary']=pd.to_numeric(df['Salary'],errors='coerce')
df['Salary']=df['Salary'].astype("Int64")
```

```python
df['Bonus']=pd.to_numeric(df['Bonus'],errors='coerce')
df['Bonus']=df['Bonus'].round().astype("Int64")
```

```python
df['Experience_Years']=pd.to_numeric(df['Experience_Years'],errors='coerce')
df['Experience_Years']=df['Experience_Years'].round().astype("Int64")
```

```python
df['Performance_Score']=pd.to_numeric(df['Performance_Score'],errors='coerce')
df['Performance_Score']=df['Performance_Score'].fillna(df['Performance_Score'].mean()).round(1)
```

```python
df['Satisfaction_Score']=pd.to_numeric(df['Satisfaction_Score'],errors='coerce')
df['Satisfaction_Score']=df['Satisfaction_Score'].fillna(df['Satisfaction_Score'].mean()).round(1)
```

```python
df['Working_Hours_Per_Week']=pd.to_numeric(df['Working_Hours_Per_Week'],errors='coerce')
df['Working_Hours_Per_Week']=df['Working_Hours_Per_Week'].round().astype('Int64')
```

```python
df['Working_Hours_Per_Week']=df['Working_Hours_Per_Week'].fillna(35)
```

```python
df['Projects_Completed']=df['Projects_Completed'].fillna(df['Projects_Completed'].mean()).round(1)
```

```python
df['Remote_Work']=df['Remote_Work'].str.capitalize()
```

```python
df['Join_Year']=pd.to_numeric(df['Join_Year'],errors='coerce')
df['Join_Year']=df['Join_Year'].astype("Int64")
```

```python
df.to_csv("clean_employee.csv")
```
