```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\WorldCupMatches.csv")
```

```python
df.info()
```

```python
df['Datetime']=pd.to_datetime(df['Datetime'],format='mixed')
```

```python
df.info()
```

```python
df['Home Team Goals']=pd.to_numeric(df['Home Team Goals'])
df['Home Team Goals']=df['Home Team Goals'].astype('Int64')
```

```python
df['Home Team Goals'].dtype
```

```python
df['Home Team Goals']
```

```python
df['Away Team Goals']=pd.to_numeric(df['Away Team Goals'],errors='coerce')
df['Away Team Goals']=df['Away Team Goals'].astype("Int64")
```

```python
df['Away Team Goals']
```

```python
df['Win conditions']=pd.to_numeric(df['Win conditions'],errors='coerce')
df['Win conditions']=df['Win conditions'].astype("Int64")
```

```python
df = df.drop('Win conditions', axis=1)
```

```python
df['Attendance']=pd.to_numeric(df['Attendance'],errors="coerce")
df['Attendance']=df['Attendance'].astype('Int64')
```

```python
df.info()
```

```python
df['RoundID']=pd.to_numeric(df['RoundID'],errors='coerce')
df['RoundID']=df['RoundID'].astype('Int64')
```

```python
df['MatchID']=pd.to_numeric(df['MatchID'],errors="coerce")
df["MatchID"]=df['MatchID'].astype("Int64")
```
